```python
#데이터 중 최고기온을 실수로 변환하여 출력
import csv
dir = '/content/drive/MyDrive/Colab Notebooks/고급파이썬/temp.csv'
f = open(dir, 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')

max_temp = -999
max_date = ''

header = next(data)
for row in data:
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    row[0] = row[0].replace('\t','')
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
f.close()
print(f'기상 관측 이래 서울의 최고 기온이 가장 높았던 날은 {max_date}로, {max_temp}도 였습니다.')
```

```python
#내 생일 기온 변화를 그래프로 그리기
import csv
import matplotlib.pyplot as plt

dir = '/content/drive/MyDrive/Colab Notebooks/고급파이썬/temp.csv'
f = open(dir, 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
next(data)
result1 = []
result2 = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '12':
            result1.append(float(row[-1]))
            result2.append(float(row[-2]))
f.close()
plt.plot(result1, 'hotpink')
plt.plot(result2, 'black')
plt.show()
```

```python
#기온 데이터를 히스토그램으로 표현
import csv
import matplotlib.pyplot as plt
dir = '/content/drive/MyDrive/Colab Notebooks/고급파이썬/temp.csv'
f = open(dir, 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
next(data)
aug = []
jan = []
for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```
